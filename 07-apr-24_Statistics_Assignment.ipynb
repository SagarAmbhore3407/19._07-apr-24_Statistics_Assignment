````python
# (i) Write a Python function to calculate the mean of a given list of numbers.
#     Create a function to find the median of a list of numbers.


def calculate_mean(numbers):
    
    if not numbers:
        return None
    return sum(numbers) / len(numbers)


def calculate_median(numbers):
   
    if not numbers:
        return None
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    mid = n // 2

    if n % 2 == 0:  # Even number of elements
        return (sorted_numbers[mid - 1] + sorted_numbers[mid]) / 2
    else:  # Odd number of elements
        return sorted_numbers[mid]

numbers = [10, 20, 30, 40, 50]

mean = calculate_mean(int_list)
median = calculate_median(int_list)

print("Mean:", mean)
print("Median:", median)
````

````python
# (ii) Develop a program to compute the mode of a list of integers


from collections import Counter

def calculate_mode(numbers):
    
    if not numbers:
        return None
    
    # Count the frequency of each number in the list
    frequency = Counter(numbers)
    
    # Find the highest frequency
    max_frequency = max(frequency.values())
    
    # Extract the mode(s)
    modes = [key for key, value in frequency.items() if value == max_frequency]
    
    return modes



modes = calculate_mode(int_list)

print("Mode(s):", modes)
````

````python
# (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

def calculate_weighted_mean(values, weights):
    
    if not values or not weights or len(values) != len(weights):
        raise ValueError("Values and weights must be non-empty lists of the same length.")
    
    # Calculate the weighted sum
    weighted_sum = sum(value * weight for value, weight in zip(values, weights))
    
    # Calculate the sum of the weights
    total_weight = sum(weights)
    
    # Calculate the weighted mean
    weighted_mean = weighted_sum / total_weight
    
    return weighted_mean


weights = np.random.randint(20,120,100).tolist()


weighted_mean = calculate_weighted_mean(int_list, weights)

print("Weighted Mean:", weighted_mean)

````

````python

# (iv) Write a Python function to find the geometric mean of a list of positive numbers


import math

def geometric_mean(nums):
    
    if not nums:
        return None
    for num in nums:
        if num <= 0:
            return None
    
    product = 1
    for num in nums:
        product *= num
    
    return product ** (1/len(nums))


geo_mean = geometric_mean(int_list)
print(f"The geometric mean of {int_list} \nis {geo_mean}")
````

````python
# (v) Create a program to calculate the harmonic mean of a list of values.

def harmonic_mean(nums):
    # Ensure the list is not empty and does not contain zero
    if not nums or 0 in nums:
        return None

    n = len(nums)
    reciprocal_sum = sum(1 / num for num in nums)
    
    return n / reciprocal_sum


harmonic_mean_value = harmonic_mean(int_list)
print(f"The harmonic mean of {int_list} \nis {harmonic_mean_value}")
````

````python
# (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of 
# outliers

import numpy as np

def trimmed_mean(nums, percentage):
    # Ensure the list is not empty and the percentage is valid
    if not nums or percentage < 0 or percentage >= 50:
        return None
    
    nums_sorted = sorted(nums)
    n = len(nums)
    k = int(n * (percentage / 100))
    
    trimmed_list = nums_sorted[k:n-k]
    
    return np.mean(trimmed_list)



percentage = 10  # Trim 10% from each end
trimmed_mean_value = trimmed_mean(int_list, percentage)
print(f"The trimmed mean of List: {int_list}\nwith Percentage: {percentage}% \nOutliers removed is: {trimmed_mean_value}")
````

````python

# (iv) Implement a function to compute the interquartile range (IQR) of a list of values.


import statistics as stats

def calculate_iqr_stats(nums):
    if len(nums) < 4:
        return None
    
    nums_sorted = sorted(nums)
    q1 = stats.median(nums_sorted[:len(nums_sorted)//2])
    q3 = stats.median(nums_sorted[(len(nums_sorted)+1)//2:])
    
    iqr = q3 - q1
    return iqr

iqr_value = calculate_iqr_stats(int_list2)
print(f"Interquartile Range (IQR): {iqr_value}")
````

````python
# (v) Build a program to calculate the coefficient of variation for a dataset.


import numpy as np

def calculate_coefficient_of_variation(nums):
    if len(nums) == 0:
        return None
    
    nums_array = np.array(nums)
    mean = np.mean(nums_array)
    std_dev = np.std(nums_array, ddof=0)  
    
    if mean == 0:
        return None
    
    cv = (std_dev / mean) * 100
    return cv


cv_value = calculate_coefficient_of_variation(int_list2)
print(f"Coefficient of Variation (CV): {cv_value:.2f}%")
````

````python

# (vi) Write a Python function to find the mean absolute deviation (MAD) of a list of numbers.

import numpy as np

def calculate_mad(nums):
    if len(nums) == 0:
        return None
    
    nums_array = np.array(nums)
    mean = np.mean(nums_array)
    abs_deviations = np.abs(nums_array - mean)
    mad = np.mean(abs_deviations)
    
    return mad


mad_value = calculate_mad(int_list2)
print(f"Mean Absolute Deviation (MAD): {mad_value:.2f}")
````

````python

#Q3.  Write a Python class representing a discrete random variable with methods to calculate its expected 
# value and variance


class DiscreteRandomVariable:
    def __init__(self, outcomes, probabilities):
    
        if len(outcomes) != len(probabilities):
            raise ValueError("The length of outcomes and probabilities must be the same.")
        if any(p < 0 or p > 1 for p in probabilities):
            raise ValueError("Probabilities must be between 0 and 1.")
        if not (abs(sum(probabilities) - 1) < 1e-6):
            raise ValueError("The sum of probabilities must be 1.")
        
        self.outcomes = outcomes
        self.probabilities = probabilities

    def expected_value(self):
        return sum(o * p for o, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        mean = self.expected_value()
        return sum(p * (o - mean) ** 2 for o, p in zip(self.outcomes, self.probabilities))



outcomes = [1, 2, 3, 4]
probabilities = [0.1, 0.2, 0.3, 0.4]

rv = DiscreteRandomVariable(outcomes, probabilities)

print(f"Expected Value: {rv.expected_value()}")
print(f"Variance: {rv.variance()}")
````

````python
#Q4.  Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and 
#     variance of the outcomes.


class FairSixSidedDie:
    def __init__(self):
        # Possible outcomes of a fair six-sided die
        self.outcomes = [1, 2, 3, 4, 5, 6]
        # Probability of each outcome (equal probability for a fair die)
        self.probabilities = [1/6] * 6

    def expected_value(self):
        return sum(o * p for o, p in zip(self.outcomes, self.probabilities))

    def variance(self):
        """
        Calculate the variance of rolling the die.
        
        :return: Variance of the outcomes.
        """
        mean = self.expected_value()
        return sum(p * (o - mean) ** 2 for o, p in zip(self.outcomes, self.probabilities))


die = FairSixSidedDie()
expected_value = die.expected_value()
variance = die.variance()

print(f"Expected Value: {expected_value:.2f}")
print(f"Variance: {variance:.2f}")
````

````python
#Q5.  Create a Python function to generate random samples from a given probability distribution (e.g., 
#     binomial, Poisson) and calculate their mean and variance.

import numpy as np

def generate_samples_and_stats(distribution, **params):

    if distribution == 'binomial':
        n = params.get('n', 10)
        p = params.get('p', 0.5)
        size = params.get('size', 1000)
        samples = np.random.binomial(n, p, size)
    
    elif distribution == 'poisson':
        lam = params.get('lam', 5)
        size = params.get('size', 1000)
        samples = np.random.poisson(lam, size)
    
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")
    
    mean = np.mean(samples)
    variance = np.var(samples)
    
    return mean, variance

# Example usage
distribution_type = 'binomial'
params = {'n': 10, 'p': 0.5, 'size': 1000}
mean, variance = generate_samples_and_stats(distribution_type, **params)
print(f"Distribution: {distribution_type}")
print(f"Mean: {mean:.2f}")
print(f"Variance: {variance:.2f}")

distribution_type = 'poisson'
params = {'lam': 5, 'size': 1000}
mean, variance = generate_samples_and_stats(distribution_type, **params)
print(f"Distribution: {distribution_type}")
print(f"Mean: {mean:.2f}")
print(f"Variance: {variance:.2f}")

````

````python
#Q6.  Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute 
#    the mean, variance, and standard deviation of the samples.


import numpy as np

def generate_gaussian_samples(mean, std_dev, size):

    # Generate random samples from the normal distribution
    samples = np.random.normal(loc=mean, scale=std_dev, size=size)
    
    # Compute mean, variance, and standard deviation of the samples
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev


mean = 0          # Mean of the normal distribution
std_dev = 1       # Standard deviation of the normal distribution
size = 1000       # Number of samples

sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(mean, std_dev, size)

print(f"Generated Gaussian Distribution Samples:")
print(f"Mean: {sample_mean:.2f}")
print(f"Variance: {sample_variance:.2f}")
print(f"Standard Deviation: {sample_std_dev:.2f}")
````

````python
# (iii) Write a function that calculates the covariance between two columns.


import numpy as np




def calculate_covariance_np(data, column1, column2):
    
    if column1 not in data.columns or column2 not in data.columns:
        raise ValueError(f"Columns '{column1}' or '{column2}' do not exist in the dataset.")
    
    # Drop missing values and extract columns
    col1_data = data[column1].dropna()
    col2_data = data[column2].dropna()
    
    # Calculate covariance using numpy
    covariance = np.cov(col1_data, col2_data, ddof=0)[0, 1]
    
    return covariance



# Calculate covariance between 'total_bill' and 'tip' using numpy
covariance_np = calculate_covariance_np(tips, 'total_bill', 'tip')
print(f"Covariance between 'total_bill' and 'tip': {covariance_np:.2f}")


````

````python
#Q11.  A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors 
#     who make a purchase). They collect data from the old and new layouts to compare.
#     To generate the data use the following command:
'''
        ```python
        import numpy as np
        # 50 purchases out of 1000 visitors
        old_layout = np.array([1] * 50 + [0] * 950)
        # 70 purchases out of 1000 visitors  
        new_layout = np.array([1] * 70 + [0] * 930)
        ```
        Apply z-test to find which layout is successful.   
'''


import numpy as np
from scipy import stats
#50 purchases out of 1000 visitors
old_layout = np.array([1] * 50 + [0] * 950)
# 70 purchases out of 1000 visitors  
new_layout = np.array([1] * 70 + [0] * 930)
print("Generated Data:")
print("Old Layout:",old_layout)
print("\nNew Layout:",new_layout)


# Calculate the conversion rates
p1 = np.mean(old_layout)  # Conversion rate for old layout
p2 = np.mean(new_layout)  # Conversion rate for new layout

# Number of trials
n1 = len(old_layout)
n2 = len(new_layout)

# Pooled proportion
p_pool = (np.sum(old_layout) + np.sum(new_layout)) / (n1 + n2)

# Standard error of the difference
se = np.sqrt(p_pool * (1 - p_pool) * (1/n1 + 1/n2))

# Z-statistic
z = (p2 - p1) / se

# P-value for one-tailed test (since we are testing if new_layout is better)
p_value = 1 - stats.norm.cdf(z)

# Print results
print(f"Conversion rate for old layout: {p1:.4f}")
print(f"Conversion rate for new layout: {p2:.4f}")
print(f"Z-statistic: {z:.4f}")
print(f"P-value: {p_value:.4f}")

# Interpret the result
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis: The new layout significantly increases the conversion rate.")
else:
    print("Fail to reject the null hypothesis: There is no significant increase in conversion rate with the new layout.")
````

````python
#Q15. A company is testing two different website layouts to see which one leads to higher click-through rates. 
#    Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of 
#    freedom, and p-value.
'''
 Use the following data:
 ```python
 layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
 layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
'''

import numpy as np
from scipy import stats
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]
print("Generated Data:")
print("layout_a_clicks:",layout_a_clicks)
print("layout_b_clicks:",layout_b_clicks)


def ab_test_analysis(layout_a_clicks, layout_b_clicks):
    # Convert lists to numpy arrays
    layout_a = np.array(layout_a_clicks)
    layout_b = np.array(layout_b_clicks)
    
    # Calculate sample statistics for Layout A
    mean_a = np.mean(layout_a)
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    n_a = len(layout_a)
    
    # Calculate sample statistics for Layout B
    mean_b = np.mean(layout_b)
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation
    n_b = len(layout_b)
    
    # Calculate t-statistic
    se = np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))
    t_statistic = (mean_a - mean_b) / se
    
    # Calculate degrees of freedom using Welch's formula
    df = ((std_a**2 / n_a + std_b**2 / n_b)**2 /
          ((std_a**2 / n_a)**2 / (n_a - 1) + (std_b**2 / n_b)**2 / (n_b - 1)))
    
    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Print results
    print(f"\nMean of Layout A: {mean_a:.4f}")
    print(f"Standard deviation of Layout A: {std_a:.4f}")
    print(f"Mean of Layout B: {mean_b:.4f}")
    print(f"Standard deviation of Layout B: {std_b:.4f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of freedom: {df:.2f}")
    print(f"P-value: {p_value:.4f}")
    
    # Interpret the result
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: There is a significant difference in click-through rates between Layout A and Layout B.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in click-through rates between Layout A and Layout B.")


ab_test_analysis(layout_a_clicks, layout_b_clicks)


````

````python

#Q16. A pharmaceutical company wants to determine if a new drug is more effective than an existing drug in 
#     reducing cholesterol levels. Create a program to analyze the clinical trial data and calculate the 
# t- statistic and p-value for the treatment effect.

'''
Use the following data of cholestrol level:
 ```python
 existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
 new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]
 ```
'''

import numpy as np
from scipy import stats
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]
print("Generated Data:")
print("existing_drug_levels:",existing_drug_levels)
print("new_drug_levels:",new_drug_levels)


def analyze_cholesterol_levels(existing_drug_levels, new_drug_levels):

    existing_drug = np.array(existing_drug_levels)
    new_drug = np.array(new_drug_levels)
    

    mean_existing = np.mean(existing_drug)
    std_existing = np.std(existing_drug, ddof=1)  # Sample standard deviation
    n_existing = len(existing_drug)
    

    mean_new = np.mean(new_drug)
    std_new = np.std(new_drug, ddof=1)  # Sample standard deviation
    n_new = len(new_drug)
    
    # Calculate t-statistic
    se = np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))
    t_statistic = (mean_existing - mean_new) / se
    
   
    df = ((std_existing**2 / n_existing + std_new**2 / n_new)**2 /
          ((std_existing**2 / n_existing)**2 / (n_existing - 1) + (std_new**2 / n_new)**2 / (n_new - 1)))
    
    # Calculate p-value for two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    # Print results
    print(f"\nMean cholesterol level (Existing Drug): {mean_existing:.2f}")
    print(f"Standard deviation (Existing Drug): {std_existing:.2f}")
    print(f"Mean cholesterol level (New Drug): {mean_new:.2f}")
    print(f"Standard deviation (New Drug): {std_new:.2f}")
    print(f"T-statistic: {t_statistic:.4f}")
    print(f"Degrees of freedom: {df:.2f}")
    print(f"P-value: {p_value:.4f}")
    
    # Interpret the result
    alpha = 0.05
    if p_value < alpha:
        print("Reject the null hypothesis: The new drug has a significantly different effect on cholesterol levels compared to the existing drug.")
    else:
        print("Fail to reject the null hypothesis: There is no significant difference in the effect of the new drug compared to the existing drug.")

analyze_cholesterol_levels(existing_drug_levels, new_drug_levels)

````

````python
#Q21. A political analyst wants to determine if there is a significant association between age groups and voter 
#    preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify 
#   them into different age groups and candidate preferences. Perform a Chi-Square test to determine if 
#   there is a significant association between age groups and voter preferences.
#    Use the below code to generate data:
'''
 ```python
 np.random.seed(0)
 age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
 voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
'''
import numpy as np
import pandas as pd
import scipy.stats as stats

np.random.seed(0)
age_groups = np.random.choice(['18-30', '31-50', '51+', '51+'], size=30)
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)
print("Generated Data:")
print("age_groups:",age_groups)
print("voter_preferences:",voter_preferences)


# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2, p_value, dof, expected = stats.chi2_contingency(contingency_table)

# Results
print(f"\nChi-Square Statistic: {chi2}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant association between age groups and voter preferences.")
else:
    print("There is no significant association between age groups and voter preferences.")


````

````python
#Q24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The 
#    company wants to determine if there is a significant difference in customer satisfaction scores among the 
#    three product versions. They conducted a survey and collected customer satisfaction scores for each 
#    version from a random sample of customers. Perform an ANOVA test to determine if there is a significant 
#    difference in customer satisfaction scores.
#    Use the following data:
'''
 ```python
 # Sample data: Customer satisfaction scores for each product version
 standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
 ```
 premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
 deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
'''

np.random.seed(0)
standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
print("Generated Data:")
print("standard_scores: ",standard_scores)
print("premium_scores: ",premium_scores)
print("deluxe_scores: ",deluxe_scores)



f_stat, p_value = stats.f_oneway(standard_scores, premium_scores, deluxe_scores)

# Results
print(f"\nF-statistic: {f_stat}")
print(f"P-value: {p_value}")

if p_value < 0.05:
    print("There is a significant difference in customer satisfaction scores among the three product versions.")
else:
    print("There is no significant difference in customer satisfaction scores among the three product versions.")


````
